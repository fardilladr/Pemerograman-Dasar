```python
print("=" * 35)
print(" SELAMAT DATANG DI TEJANO BLUE")
print("=" * 35)
print("Menu Minuman & Dessert:")
print("1. Midnight Caramel           - Rp 32.000")
print("2. Scarlet Blue Cakey         - Rp 24.000")
print("3. Nebula Crème Brûlée        - Rp 35.000")
print("4. Aqua Teac Splash           - Rp 17.000")
print("5. Sea Wave Cheesecake        - Rp 26.000")
print("6. Wave Crumble Strawberry    - Rp 29.000")
print("7. Tejano Dark Lotus          - Rp 38.000")
print("=" * 35)

# Input pilihan menu
pilihan = int(input("Silakan pilih nomor menu (1-7): "))

# Menentukan nama dan harga berdasarkan pilihan
if pilihan == 1:
    nama = "Midnight Caramel"
    harga = 32000
elif pilihan == 2:
    nama = "Scarlet Blue Cakey"
    harga = 24000
elif pilihan == 3:
    nama = "Nebula Crème Brûlée"
    harga = 35000
elif pilihan == 4:
    nama = "Aqua Teac Splash"
    harga = 17000
elif pilihan == 5:
    nama = "Sea Wave Cheesecake"
    harga = 26000
elif pilihan == 6:
    nama = "Wave Crumble Strawberry"
    harga = 29000
elif pilihan == 7:
    nama = "Tejano Dark Lotus"
    harga = 38000
else:
    print("Maaf, pilihan menu tidak valid.")
    exit()

# Input jumlah pesanan
jumlah = int(input(f"Masukkan jumlah {nama}: "))
total = harga * jumlah

# Menentukan diskon
if total >= 200000:
    diskon = 0.70
elif total >= 100000:
    diskon = 0.50
elif total >= 50000:
    diskon = 0.30
else:
    diskon = 0

# Menghitung potongan dan total bayar
potongan = total * diskon
total_bayar = total - potongan

print("-" * 35)
print(f"Pesanan Anda   : {nama}")
print(f"Jumlah         : {jumlah}")
print(f"Harga Satuan   : Rp {harga:,}")
print(f"Total Belanja  : Rp {total:,}")

# Jika total kurang dari 20000, tidak menampilkan total bayar
if total < 20000:
    print("Diskon         : Tidak ada")
else:
    if diskon > 0:
        print(f"Diskon         : {int(diskon * 100)}%")
        print(f"Potongan Harga : Rp {int(potongan):,}")
    else:
        print("Diskon         : Tidak ada")
    print(f"Total Bayar    : Rp {int(total_bayar):,}")

print("-" * 35)

# Pembayaran hanya jika total >= 20000
if total >= 20000:
    uang = int(input("Masukkan jumlah uang Anda: Rp "))
    if uang >= total_bayar:
        kembali = uang - total_bayar
        print(f"Kembalian Anda : Rp {int(kembali):,}")
        print("Terima kasih telah berkunjung ke Tejano Blue!")
    else:
        kurang = total_bayar - uang
        print(f"Maaf, uang Anda kurang Rp {int(kurang):,}")
else:
    print("Thank you, enjoy at Tejano Blue!")
```

```python

print("^" * 70)
print("                           Tejano Blue     ")
print("^" * 70)

print("Our Signature Menu : ")
print("1. Midnight Caramel                 - Rp 32.000")
print("2. Scarlet Blue Cakey               - Rp 24.000")
print("3. Nebula Crème Brûlée              - Rp 35.000")
print("4. Aqua Teac Splash                 - Rp 17.000")
print("5. Sea Wave Cheesecake              - Rp 26.000")
print("6. Wave Crumble Strawberry          - Rp 29.000")
print("7. Tejano Dark Lotus                - Rp 38.000")
print("-" * 70)

# Daftar menu
menu = {
    1: ("Midnight Caramel", 32000),
    2: ("Scarlet Blue Cakey", 24000),
    3: ("Nebula Crème Brûlée", 35000),
    4: ("Aqua Teac Splash", 17000),
    5: ("Sea Wave Cheesecake", 26000),
    6: ("Wave Crumble Strawberry", 29000),
    7: ("Tejano Dark Lotus", 38000)}

# Input jumlah item
pesanan_list = []
total_harga = 0
jumlah_item = int(input("Ada berapa menu yang ingin anda pesan? "))

for i in range(jumlah_item):
    pilihan = int(input(f"Pilih menu ke-{i+1} (1-7): "))
    if pilihan in menu:
        nama, harga = menu[pilihan]
        jumlah = int(input(f">Masukkan jumlah untuk {nama}: "))
        subtotal = harga * jumlah
        pesanan_list.append((nama, jumlah, harga, subtotal))
        total_harga += subtotal
        print(f" {jumlah}x {nama} ditambahkan (Subtotal: Rp {subtotal:,})")
    else:
        print("Menu tidak tersedia, silahkan pilih menu lain.")

# Hitung diskon
if total_harga >= 200000:
    diskon = 0.70
elif total_harga >= 100000:
    diskon = 0.50
elif total_harga >= 50000:
    diskon = 0.30
else:
    diskon = 0

potongan = total_harga * diskon
total_bayar = total_harga - potongan

# Cetak struk
print("^" * 70)
print("                            Tejano Blue     ")
print("^" * 70)

print(f"{'Item':<35}{'Qty':<8}{'Harga per Item':<18}{'Subtotal'}")
print("-" * 70)

for nama, jumlah, harga, subtotal in pesanan_list:
    print(f"{nama:<35}{jumlah:<8}Rp {harga:<15,}Rp {subtotal:,}")

print("-" * 70)
print(f"{'Total Belanja':<61}Rp {total_harga:,}")

if diskon > 0:
    print(f"{'Diskon':<61}{int(diskon*100)}%")
    print(f"{'Potongan':<61}Rp {potongan:,.0f}")
else:
    print(f"{'Diskon':<61}Tidak ada")

print(f"{'Total Bayar':<61}Rp {total_bayar:,.0f}")
print("-" * 70)

uang_dibayar = int(input("Bayar                      : Rp "))
if uang_dibayar >= total_bayar:
    kembalian = uang_dibayar - total_bayar
    print(f"{'Kembalian':<61}Rp {kembalian:,.0f}")
else:
    kekurangan = total_bayar - uang_dibayar
    print(f"Maaf, Pembayaran anda kurang sebesar Rp {kekurangan:,.0f}.")
    print("Harap selesaikan pembayaran anda.")


print("^" * 70)
print("                   Thank You, Enjoy at Tejano Blue!      ")
print("^" * 70)
```
